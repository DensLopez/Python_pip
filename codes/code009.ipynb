# ¿Qué es un ambiente virtual?

Por ejemplo nosotros tenemos un ambiente con matplotlib 3.6 de manera global, con ciertos módulos de numpy y pandas. A esto se le llama que se instalan de forma o general o global, y esto significa que estas librerías quedan a uso compartido para cualquiera que necesite ejecutarlo en la maquina.



Esto puede generar un poco de conflictos porque si tienes diferentes proyectos o estas con diferentes compañías, microservicios pueden empezar a chocar o tener conflictos entre ellos



Digamos que para cada proyecto como los que se ven amarillo, necesitas ciertas versiones en especifico de cada librería u otro tipo de librerías o módulos.

Al tener todo de forma compartida, puede empezar a colapsar y a generar errores.

## Ambientes virtuales
Para esta problemática se crearon los ambientes virtuales en Python.

Los ambientes virtuales en Python encapsulan cada uno de los módulos y lo atan a cada proyecto, y no lo dejan en una zona compartida.

Esto quiere decir que cada proyecto tiene sus propias dependencias, sin que tengan conflictos con otros proyectos. 

**LO QUE TENEMOS DE FORMA COMPARTIDA ES LA VERSION DE PYTHON 3**
