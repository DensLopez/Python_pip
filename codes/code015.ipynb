# ¿Qué es [docker](https://docs.docker.com/)?

Docker es una herramienta que también nos sirve para aislar entornos.

Recapitulemos de cómo hemos aislado nuestros proyectos

1. Primero teníamos una versión compartida de Python y módulos que se podía enlazar o comunicar entre sí.



2. Luego, con los entornos virtuales de Python logramos aislar cada dependencia y cada modulo a ca proyecto pero conservando o teniendo en común a  Python como una versión global, con esto logramos que no exista un choque entre proyectos y sus dependencias. Este tipo de ambiente es muy común en Ciencia de Datos y algunas otras areas.



3. Cuando ya necesitamos llevar el código a producción y servidores en la nube nos encontramos con un problema, y es que también necesitamos aislar la versión de Python. Porque también nos podemos encontrar con proyectos en los cuales las versiones de Python son diferentes. Por ejemplo podemos tener un proyecto con Python 2.7, otro con Python 3.8 o Python 3.10. Entonces lo que hace Docker es **aislar** no solo las dependencias, sino también el entorno de ejecución y con ello las versiones que estamos corriendo de Python.



Esto lo hace con **contenedores**, una tecnología que posee una gran característica que es; así como funciona la aplicación en mi computadora, se puede realizar un despliegue para que funcione en cualquier equipo; ya sea en la nube (Amazon. Google Cloud) o local.

Un ejemplo que podríamos desplegar es un web server en la nube. En las próximas clases veremos como funciona este tipo de tecnología.

> Aquí tienes 5 razones por las que Docker es esencial:

- **Portabilidad y Consistencia:** Docker encapsula una aplicación junto con todas sus dependencias, bibliotecas, y configuraciones en un contenedor. Esto garantiza que la aplicación funcionará de la misma manera en cualquier entorno, ya sea en un equipo de desarrollo, un servidor de pruebas o en producción, eliminando problemas de "funciona en mi máquina".
- **Aislamiento de Aplicaciones:** Docker permite ejecutar múltiples aplicaciones en contenedores separados en la misma máquina, sin que interfieran entre sí. Cada contenedor tiene su propio entorno, lo que evita conflictos entre versiones de software o dependencias.
- **Despliegue Rápido y Escalabilidad:** Docker permite desplegar aplicaciones rápidamente y escalar horizontalmente sin complicaciones. Puedes lanzar nuevas instancias de una aplicación en contenedores en cuestión de segundos, lo que facilita la respuesta a cambios en la demanda.
- **Eficiencia de Recursos:** Los contenedores de Docker son ligeros y utilizan menos recursos que las máquinas virtuales tradicionales porque comparten el mismo sistema operativo base. Esto permite que más contenedores se ejecuten en la misma infraestructura, reduciendo los costos de hardware y mejorando la eficiencia.
- **Integración con CI/CD:** Docker se integra perfectamente con flujos de trabajo de integración continua y despliegue continuo (CI/CD). Permite automatizar la construcción, prueba y despliegue de aplicaciones de manera confiable y reproducible, acelerando el ciclo de desarrollo y mejorando la calidad del software.
