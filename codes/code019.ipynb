# Dockerizando web services

En las clases anteriores Dockerizamos un script, si funciona pero el uso más común es para tener un sistema encendido y ademas se utiliza mucho para desarrollo web.

## Dockerizar web server

Vamos a copiar los archivos [Dockerfile](py-project/app/Dockerfile) y [docker-compose.yml](py-project/app/docker-compose.yml) a la carpeta [web-server](py-project/web-server/).

Todo esto lo hacemos de forma manual.
- [Dockerfile](py-project/web-server/Dockerfile).
- [docker-compose.yml](py-project/web-server/docker-compose.yml).

Ahora procedemos a modificar los archivos anteriores para que queden listos para su aplicación 

#### **Construir contenedor**

Primero nos posicionamos en la carpeta `web-server` y procedemos a correr los comandos siguientes:

```bash
#Entrando a la carpeta web-server
cd web-server
#Construir el contenedor
docker-compose build
#Levantar el contenedor
docker-compose up -d
#Ahora veamos el estado del contenedor
docker-compose ps
```

Una vez que nos aparece lo siguiente



Vamos al navegador y hacemos lo siguiente:

- En la barra de búsqueda ponemos: `localhosts`



- En la barra de búsqueda ponemos: `http://localhost/contact`



- En la barra de búsqueda ponemos: `http://localhost/html`



Como se puede ver hemos logrado Dockerizar una aplicación web.
