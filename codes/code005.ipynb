# Python con Git y Git Hub

Vamos a unir nuestro repositorio local a un repositorio de Git Hub.

Iniciamos nuestra cuenta en Git Hub, vamos a repositories y despus seleccionamos el botn `New repository`. Elegimos nuestra cuenta personal en el campo `owner`.Y en **Repository name** ponemos curso-python-pip. Ponemos descripcin, lo dejamos en pblico y damos en `Create repository`.

### Conectarlo al repositorio local
Para conectar regresamos a la terminal y procedemos a ejecutar las lineas siguientes:

```git
git remote add origin enlace_gitgub
```



Recuerda que antes de agregar el repositorio remoto a tu repositorio local, debemos tener uno. Entonces recuerda haber corrido el comando:
`git init`. Se debe crear una carpeta con el nombre `.git`, que es donde se guardan todos los cambios que tu realizas en tu repositorio.



Una vez que se haya creado ahora si procedemos a ejecutar el comando de aadir repositorio

```bash
git remote add origin enlace_gitgub

git remote -v #Nos dice la direccin del repositorio remoto
```


Ahora, tengo estos comando para ejecutar operaciones en Git y enlazar mi repositorio remoto:

```bash
git add --all 	#Para aadir todo, inclusive archivos en diferentes carpetas

git commit -m "Mensaje de commit"	#Para hacer commit

git commit -am "Mensaje de commit"   #Para hacer aadir cambios al stage y hacer commit

git pull origin main	#Trae los cambios o archivos de tu repositorio remoto

git pull origin main --allow-unrelated-histories	#Forzar el merge de repositorios no relacionados

git push origin main		#Envia los cambios al repositorio remoto

git log		#Ver el historial del repositorio junto a sus commits y ID's

gitk	#Me arroja una ventana con datos visuales del repositorio
```

Entonces, para realizar el primer commit y mandarlo al repositorio remoto quedar como;

```bash
git add --all
git commit -m "Primera versin del proyecto"
git push origin main
```

Luego vamos a nuestro repositorio remoto y vemos los cambios.



Me aparecen mi primer archivo.

Despus vamos a crear un archivo llamado: `.gitignore` que lo colocaremos en nuestro repositorio.

Para eso nos apoyamos de la herramienta [gitignore.io](https://www.toptal.com/developers/gitignore/) y procedemos a buscar "Windows, Unix, Mac y Python"

Importante: aunque no tengamos todos los sistemas operativos necesitamos poner los 3 porque puede que trabajemos con alguien que si los tenga y entonces omitimos ciertos aspectos importantes.



Despus nos creara un [archivo](https://www.toptal.com/developers/gitignore/api/macos,windows,linux,python) que vamos a copiar en un archivo [.gitignore](py-project/.gitignore) 

Recuerda que este archivo es para ignorar los archivos que no necesitamos agregar al repositorio, ya que no aportan nada a nuestro cdigo.

Tambin vamos a agregar otro archivo llamado **[README.md](py-project/README.md)**. Este es un archivo Markdown en el cual ponemos instrucciones para que las personas que quieran empezar a trabajar con este proyecto las vean, y as poder utilizarlo.

Una vez tenemos los cambios, vamos a usar git:

```bash
git add --all
git commit -m "Archivos .gitignore y README.md"
git pull origin main
git push origin main
```

Una vez los llevemos vamos a ver como qued en Git Hub

Como se puede observar ya se han subido los archivos y por defecto Git Hub me muestra mi archivo **[README.md](py-project/README.md)** como portada del repositorio. 

En este caso este es mi [path relativo](py-project)

Por fin ya tenemos los 2 archivos principales para comenzar un proyecto en Python.
- [README.md](py-project/README.md)
- [.gitignore](py-project/.gitignore)
