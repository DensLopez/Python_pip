Y para guardar la imagen, tendremos que usar la siguiente linea, para que se ejecute correctamente:

```python
def generate_bar_chart(name, labels, values):
  fig, ax = plt.subplots()
  ax.bar(labels, values)
  plt.savefig(f'./images/{name}-bar.png')
  plt.close()

def generate_pie_chart(labels, values):
  fig, ax = plt.subplots()
  ax.pie(values, labels=labels)
  ax.axis('equal')
  plt.savefig(f'./images/pie.png')
  plt.close()

```

**IMPORTANTE:** La carpeta `images` ya debe estar creado para que el archivo no arroje error.
